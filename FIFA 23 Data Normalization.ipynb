### Min Max Normalization

```python
# from sklearn.preprocessing import MinMaxScaler
```

```python
!pip list
```

```python
# !pip install scikit-learn
```

```python
# This uses an equation instead of the MinMaxScaler sklearn library
nm_age=(fifa['Age'] - fifa['Age'].min())/(fifa['Age'].max()-fifa['Age'].min())
nm_overall=(fifa['Overall'] - fifa['Overall'].min())/(fifa['Overall'].max()-fifa['Overall'].min())
nm_value=(fifa['Value(in Euro)'] - fifa['Value(in Euro)'].min())/(fifa['Value(in Euro)'].max()-fifa['Value(in Euro)'].min())
```

```python
nm_overall.describe()
```

```python
nm_age = pd.DataFrame(nm_age, columns = ['Age'])
nm_overall = pd.DataFrame(nm_overall, columns = ['Overall'])
nm_value = pd.DataFrame(nm_value, columns = ['Value(in Euro)'])
```

```python
nm_fifa=pd.concat([nm_age, nm_overall,nm_value], axis=1)
nm_fifa
```

```python
nm_fifa.describe()
```

```python
nm_fifa[['Age']].plot(kind = 'hist', bins=10,figsize=(10,6))
```

### Z-score Normalization

```python
from sklearn.preprocessing import StandardScaler
```

```python
!pip install scikit-learn
```

```python
# This uses an equation instead of the StandardScaler sklearn library
z_age=(fifa['Age'] - fifa['Age'].mean())/(fifa['Age'].std())
z_overall=(fifa['Overall'] - fifa['Overall'].mean())/(fifa['Overall'].std())
z_value=(fifa['Value(in Euro)'] - fifa['Value(in Euro)'].mean())/(fifa['Value(in Euro)'].std())
```

```python
z_age = pd.DataFrame(z_age, columns = ['Age'])
z_overall = pd.DataFrame(z_overall, columns = ['Overall'])
z_value = pd.DataFrame(z_value, columns = ['Value(in Euro)'])
```

```python
z_fifa = pd.concat([z_age,z_overall,z_value], axis=1)
z_fifa
```

```python
z_fifa.describe()
```

```python
z_fifa[['Age']].plot(kind = 'hist', bins=10,figsize=(10,6))
```
