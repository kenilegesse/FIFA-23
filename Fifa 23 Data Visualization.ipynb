# FIFA Data Visualization

```python
import pandas as pd
```

```python
fifa = pd.read_csv('C:\\Users\\kenes\\Desktop\\Documents\\Data Science\\FIFA 23 PROJECT\\Fifa 23 Players Data.csv')
fifa.info()
%matplotlib inline
```

```python
fifa.head()
```

### Project relationship visualization (scatter / line plots)

```python
fifa.plot(kind ='scatter',x ='Age',y ='Wage(in Euro)')
```

### Distribution Visualization (histogram/ symmetry and skew) 

```python
fifa['Overall'].plot(kind = 'hist')
```

```python
fifa['Age'].plot(kind = 'hist')
```

### Comparison Visualization (stacked/ grouped bar and pie charts) 

```python
fifa.pivot(columns= 'Age' , values ='Overall')
# NaN means not a number
```

```python
fifa.pivot(columns= 'Age' , values ='Overall').plot(kind='box',figsize =(10,6))
```

## Composition Visualization (stacked/ grouped bar and pie chart 

```python
fifa.groupby('Age')['Best Position'].value_counts()
```

```python
fifa.groupby('Age')['Best Position'].value_counts().unstack()
```

```python
fifa.groupby('Age')['Best Position'].value_counts().unstack().plot(kind='bar',stacked=True,figsize=(10,6))
```
