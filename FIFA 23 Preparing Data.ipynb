### Adjusting missing data

```python
import pandas as pd
fifa = pd.read_csv('C:\\Users\\klegesse\\Desktop\\New folder\\Fifa 23 Players Data.csv')
fifa
```

### Masking for missing rows

```python
mask = fifa['Value(in Euro)'].isnull() 
mask
```

```python
# fifa(mask)
# To view the complete row of the masked items
```

### Removing rows of missing data

```python
fifa.dropna()
```

### Removing rows of missing data for specific columns

```python
fifa = fifa.dropna(subset = ['Age','Potential','Value(in Euro)'], how = 'all')
fifa
```

### Cleaning columns

```python
fifa = fifa.dropna(axis=1)
fifa
```

### Tresh of dropping a column 

```python
fifa = fifa.dropna(axis=1, thresh = 10)
fifa
```

### To fill missing value

```python
fifa = fifa.fillna({'Age':'25'})
fifa
```

### To replace the missing column values with a median of that column

```python
fifa=fifa.fillna({'Age':fifa['Age'].median()})
fifa
```

 ### To replace a value on a cell by cell basis

```python
mask = (fifa['Full Name']== 'Lionel Messi') & (fifa['Age'] == 35)
fifa.loc[mask,:]
```

```python
fifa.loc[mask,'Potential'] = 40
fifa
#This process would update MESSI's age to 40 years if the main .csv has a NaN value
```
